## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Find the totals  for district

#Remove the repeated data, schdata

schdata= student_data['school_name'].unique()  

#Get the count of the schdata, schtotal

schtotal = len(schdata) 

#Get the the number of total students, sttotal
sttotal= student_data['student_name'].count() 

#summarize the total budget for schools, budt
budt = school_data['budget'].sum() 

# Find the Averages for the students' scores

# Average Math score, avmaths
# Average Reading score, avreads

avmaths=student_data['math_score'].mean() 
avreads=student_data['reading_score'].mean() 

# Locate the students who pased (GREATER THAN 70 AS STATED IN THE INSTRUCTIONS)
#Students who approbe math, passm
#Studentes who approve reading, passr

passm = student_data.loc[student_data["math_score"] >= 70, :] 
passr = student_data.loc[student_data["reading_score"] >= 70, :]

#Count the students who approved Math, countm
#Count the students who approved Reading, countr

countm = passm['math_score'].count()  
countr = passr['reading_score'].count() 

# Get the percentages of the students who approved agains the total population

#Students Percentage who approved Math, percm
#Students Percentage who approved Reading, percr

percm = (countm/sttotal)*100
percr = (countr/sttotal)*100

#Get the Overall Passing Rate for approval, both subjects combined together against the total students, ppoverall

ppoverall = (countm + countr)/sttotal*50   

# Construct the Data Frame, df1


df1 = pd.DataFrame({
    
    "Total Schools": [schtotal],
    "Total Students": [sttotal],
    "Total Budget": [budt],
    "Average Math Score": [avmaths],
    "Average Reading Score": [avreads],
    "% Passing Math":[percm],
    "% Passing Reading":[percr],
    "% Overall Passing Rate": [ppoverall]})

#Give Format to the Selected Data, with 2 decimals 

df1['Total Budget'] = df1['Total Budget'].map('${:,.2f}'.format)
df1['Average Math Score'] = df1['Average Math Score'].map('{:,.2f}'.format)
df1['Average Reading Score'] = df1['Average Reading Score'].map('{:,.2f}'.format)
df1['% Passing Math'] = df1['% Passing Math'].map('{:,.2f}'.format)
df1['% Passing Reading'] = df1['% Passing Reading'].map('{:,.2f}'.format)
df1['% Overall Passing Rate'] = df1['% Overall Passing Rate'].map('{:,.2f}'.format)

#Get the final Data Frame as requested in the Instructions

df1=df1[['Total Schools','Total Students','Total Budget',
        'Average Math Score', 'Average Reading Score','% Passing Math','% Passing Reading',
        '% Overall Passing Rate']]

df1
```

## Top Performing Schools (By Passing Rate)

```python
# To merge from schools_data as in the frame with the Average Scores(students)

#Get together de Data Frame and grouped by School name, schname

schname =student_data.groupby(['school_name']) 

#Get from the schname Data Frame the Math and reading scores according to its school, schscores

schscores = schname[['school_name','math_score','reading_score']].mean().reset_index() 

# Get the merger of the schools data with the  average scores data, schmerge
#Takes into merger Ponint the SCHOOL_NAME to match the Data Frames

schmerge = pd.merge(schscores, school_data, on='school_name') 

# Obtain the Approved students per school, then get into merge with df2, the constructed final Data Frame with the results

#Get the math and reading approvals into two groups by School Name

#Math Approved , mathapp
#Reading Approved, readapp

mathapp= passm.groupby(["school_name"]) 
readapp  = passr.groupby(["school_name"])

#Get in to the Data Frame to be prepared to merge Math and Reading Approval Rate by School Name

#Get the Data Frame  for Approval in Math Score by School Name, mathapp_df
#Get the Data Frame  for Approval in Reading Score by School Name, readapp_df

mathapp_df=mathapp[['math_score']].count().reset_index()
readapp_df = readapp[['reading_score']].count().reset_index()

#Get the new Data Frames with the Finals results by consideraing the Approval Rates for Math and Reading by School NAme with the
#whole results to get into the next two results as follows

# First, find the merge from the Data Frame Averages-schmerge(Scores & School Name)- with the Approval Rate in Math -mathapp_df-
# Stored the merger into df3
df3 = pd.merge(schmerge, mathapp_df, on='school_name')

#Now we have the First JOINT, now we proceed with this filter to match with the Reading Approval Rate

df3= pd.merge(df3, readapp_df, on='school_name')

# Now proceed to Get into df3 the Results remaining 

#Define the Size from Data Frame to get the proper Percentages

df3['% Passing Math'] = df3['math_score_y']/df3['size']*100
df3['% Passing Reading'] = df3['reading_score_y']/df3['size']*100
df3['% Overall Passing'] = (df3['% Passing Math'] + df3['% Passing Reading'])/2
df3['Per Student Budget'] = df3['budget']/df3['size']

#Get the Final Data Frame for the Final Results, df4

#Rename the Columns with the previous df3 to get the final results as estipulated:

df4 = df3.rename(index=str, columns={"school_name":"School",'math_score_x':'Average Math Score',
            'reading_score_x':'Average Reading Score','type':'School Type',
            'size':'Total Students','budget':'Total Budget'})

df4 = df4.drop(['math_score_y','reading_score_y','School ID'], axis = 1)


df4 = df4[["School",'School Type','Total Students','Total Budget','Per Student Budget',
           'Average Math Score', 'Average Reading Score','% Passing Math',
            '% Passing Reading','% Overall Passing']]

#Set the Index to be displayed as School
df4 = df4.set_index('School')

#Give Format to the Data in decimals=2

df4['Total Budget'] = df4['Total Budget'].map('${:,.2f}'.format)
df4['Per Student Budget'] = df4['Per Student Budget'].map('${:,.2f}'.format)
df4['Average Math Score'] = df4['Average Math Score'].map('{:,.2f}'.format)
df4['Average Reading Score'] = df4['Average Reading Score'].map('{:,.2f}'.format)
df4['% Passing Math'] = df4['% Passing Math'].map('{:,.2f}'.format)
df4['% Passing Reading'] = df4['% Passing Reading'].map('{:,.2f}'.format)
df4['% % Overall Passing'] = df4['% Overall Passing'].map('{:,.2f}'.format)
df4.head(10)
```

```python
#Display the top Five Schools, t5

t5 = df4.sort_values(by=["% Overall Passing"], ascending=False)
t5.head(5)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

```

```python
# Get the SCORES BY SCHOOL SPENDING

#Get the School size from df3 and apply bins function

schoolbinsspend = pd.cut(df3['Per Student Budget'], spending_bins, labels=group_names)

schoolspending = pd.DataFrame({'Spend Level':schoolbinsspend,
            'Average Math Score':schmerge['math_score'],                                       
            'Average Reading Score':schmerge['reading_score'],
            '% Passing Math':df3['% Passing Math'], 
            '% Passing Reading':df3['% Passing Reading'],
           'Overall Passing Rate':df3['% Overall Passing']})

schoolspending.sort_values('Spend Level')
grschoolspending = schoolspending.groupby(['Spend Level'])

spending_student = grschoolspending.mean()
spending_student = spending_student[['Average Math Score', 'Average Reading Score', '% Passing Math',
                   '% Passing Reading', 'Overall Passing Rate']]
spending_student
```

```python
# Get the scores by school size.Using the Size bins and group names as defined by the instructor

#Get the School size from df3 and apply bins function

schoolbins = pd.cut(df3["size"], size_bins, labels=group_names)

schoolsize = pd.DataFrame({'School Size':schoolbins,
                                        'Average Math Score':schmerge['math_score'],                                       
                                        'Average Reading Score':schmerge['reading_score'],
                                        '% Passing Math':df3['% Passing Math'], 
                                        '% Passing Reading':df3['% Passing Reading'],
                                        'Overall Passing Rate':df3['% Overall Passing']})

grschoolsize = schoolsize.groupby(['School Size'])

schoolresult = grschoolsize.mean()
schoolresult = schoolresult[['Average Math Score', 'Average Reading Score', '% Passing Math',
                                   '% Passing Reading', 'Overall Passing Rate']]
schoolresult
```
